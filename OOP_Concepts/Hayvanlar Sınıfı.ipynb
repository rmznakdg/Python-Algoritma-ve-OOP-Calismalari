```python
class Hayvan():
    def __init__(self,isim,yaş):
        self.isim = isim 
        self.yaş = yaş
        print("Hayvan sınıfının init fonksiyonu")
    def ses(self):
        print("Bu hayvan ses çıkartıyor")
    def bilgileriGoster(self):
        print(f"İsim:{isim}\nTür:Hayvan\nYaş:{yaş}")
        
class Köpek(Hayvan):
    def __init__(self,isim,yaş,cins):
        super().__init__(isim,yaş)
        self.cins = cins
    def ses1(self):
        print("Havhav")
    def bilgileriGoster(self):
        print(f"İsim:{self.isim}\nTür:Köpek\nYaş:{self.yaş}\nCins:{self.cins}")
class At(Hayvan):
    def __init__(self,isim,yaş,cins):
        super().__init__(isim,yaş)
        self.cins = cins
    def ses1(self):
        print("Ahiii")
    def bilgileriGoster(self):
        print(f"İsim:{self.isim}Tür:At\n\nYaş:{self.yaş}\nCins:{self.cins}")
class Kuş(Hayvan):
    def __init__(self,isim,yaş,cins):
        super().__init__(isim,yaş)
        print("Kuş sınıfının init fonksiyonu")
        self.cins = cins
    def ses1(self):
        print("Cikcik")
    def bilgileriGoster(self):
        print(f"İsim:{self.isim}\nTür:Kuş\nYaş:{self.yaş}\nCins:{self.cins}")
```

```python
kuş = Kuş("Kanarya",15,"Papağan")
```

```python
kuş.ses1()
```

```python
kuş.bilgileriGoster()
```
