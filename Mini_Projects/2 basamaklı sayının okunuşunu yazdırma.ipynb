```python
def okunusSayi(sayi):
    birler = ["","bir","iki","üç","dört","beş","altı","yedi","sekiz","dokuz"]
    onlar = ["","on","yirmi","otuz","kırk","elli","altmış","yetmiş","seksen","doksan"]

    birlerb = sayi % 10
    onlarb = sayi // 10

    return (onlar[onlarb] + "" + birler[birlerb])
sayi = int(input("Sayıyı giriniz:"))
print(okunusSayi(sayi))
```
