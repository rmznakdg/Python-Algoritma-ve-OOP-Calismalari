```python
toplam = 0
girilenSayılar = []
i = 0
while True:
    sayi = input("Sayı giriniz:(Çıkmak için 'q' tuşuna basın)")
    if sayi == "q":
        break
    else:
        toplam += int(sayi)
        girilenSayılar.append(sayi)
        print("{}. girilen sayı:{}".format(i+1,girilenSayılar[i]))
        i+=1
print("toplam:",toplam)
print(girilenSayılar)
    
                
```
