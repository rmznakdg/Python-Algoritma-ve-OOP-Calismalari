```python
sayi = int(input("Bir sayı giriniz:"))
toplam = 0
i = 1
while i < sayi:
    if sayi % i == 0:
        toplam+=i
        i+=1
    else:
        i+=1    
if sayi == toplam:
    print(f"{sayi} sayısı bir mükemmel sayıdır.")
else:
    print("Bu sayı mükemmel değildir")
```
